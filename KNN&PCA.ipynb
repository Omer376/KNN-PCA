#**Q1)What is K-Nearest Neighbors (KNN) and how does it work?**
#ANS:-
KNN is a supervised machine learning algorithm which is used for both classification and regression problem. Its basic idea is we inherit property from the people surronding us. It works on the basis of the following steps below:-
* We firstly plot the datapoints in n-d space
* We initialize the value of K(idealy it should be greater than 3 and always greater than 0)
* For each value of K, we calculate the distance of new datapoint from other datapoint in the n-d space. We calculate them by either using Euclidean distance formula or Manhatta distance. We then sort  the values and based on k we find the class of the new datapoint.
* We do that getting mode of classes for each nearest datapoint and assign it to the new datapoint as prediction.
* In case of regression, we take the mean of all the nearest datapoint values and give it as a prediction. Rest the steps are the same.

#**Q2) What is the difference between KNN Classification and KNN Regression?**
#ANS:-
Although the steps are similar for both classification and regression problems using KNN, the diffrence arises at the time of prediction.
* In classifier, we take the mode of the classes from nearest datapoint to the new datapoint based on K. The most voted class from the nearest neighbours will be predicted as class of new datapoint.

* In regression, we take the mean from the continuos value of the nearest neighbours. This mean will be assigned as predicted value of the new datapoint.

#**Q3)What is the role of the distance metric in KNN?**
#ANS:-
The role of distance in KNN is to compute the nearest neighbours aka datapoints to the new datapoint. Then based on K, we will know the nearest datapoints to the test data. We calculate distance in 2 ways:-

* Euclidean distance
* Manhatta distance

#**Q4)What is the Curse of Dimensionality in KNN?**
#ANS:-
Curse of dimensionality is a problem in which with increasing no.of features, at one point the increase in accuracy will be very low in such a way you can't tell a see massive diffrence in accuracy. It can be due to the following reasons:-
* A feature may be reapeated or duplicated.
* SOme features may be multicollinear.
* AFter some time the variance or information gain will be nill.
* Lots of unnesscary features present in data.

#**Q6)What are KD Tree and Ball Tree in KNN?**
#ANS:-
KNN is a computationally expensive algorithm as we calculate the distance for each value of K, which may be expensive as the no.of features increase. KD tree and Ball tree are variants of KNN to overcome computational time:-

1. KD Tree(K-Dimensional): It's main idea is to partition data into a binary tree(a root has no more than 2 childs) and implment the steps which are given below:-
* Sort each features in the data.
* Calculate the median for each of the feature. We usually take a integer value of median as float value may not be present in the data which will make splitting impossible.
* Now we partition the data recursively for each median , it means we create sub partition also based on median.
* Now we finally construct the KD tree where we place the splitting pattern which we did based on median.
* It's main advantage is that we do not need to calculate distance, we can simple check the KD tree to find nearest datapoints. It will look like,

2. Ball Tree:- It is easier comparing to KD Tree and involves the following steps:-
* Each datapoint will be group/ball with itself.
* Now we keep combining nearest datapoints recursively until the end result is one singular ball.
* For test data we only calculate the distance from its won ball or nearest ball, the end result will look something like this

#**Q7) When should you use KD Tree vs. Ball Tree?**
#ANS:-
1. Use KD Tree when:
* Dimensions are low (< 2030).
* Distance metric is Euclidean (L2) or Manhattan (L1).
* Data is numeric and axis-aligned splits are fine.

2. Use Ball Tree when:
* Dimensions are higher (> 20).
* Custom or non-Euclidean distance metrics are needed.
* Data distribution is irregular or not well-suited to axis-aligned splits.

#**Q8)What are the disadvantages of KNN?**
#ANS:-
* Slow prediction speed  must compute distance to all training points for each query.
* High memory usage  needs to store the entire training dataset.
* Sensitive to irrelevant/noisy features  irrelevant features can distort distance calculations.
* Performance drops in high dimensions  suffers from the curse of dimensionality.
* Needs feature scaling  inconsistent units can bias distance measurement.
* No learning phase  doesnt build a model, so cant generalize or compress data well.
* Sensitive to outliers  outliers can mislead nearest neighbor selection.

#**Q9)How does feature scaling affect KNN?**
#ANS:-
Prevents bias toward features with larger ranges  without scaling, features with bigger numeric values dominate distance calculations.
Ensures fair contribution of all features  scaling puts all features on the same scale (e.g., 01 or mean=0, std=1).Improves distance accuracy  KNN relies on distance metrics like Euclidean/Manhattan, which are sensitive to magnitude differences.Leads to better neighbor selection  avoids wrong nearest neighbors caused by unscaled data.

#**Q15)How do you decide the number of components to keep in PCA?**
#ANS:-
* Choose components that keep 9095% variance.
* Use Scree Plot to find where the curve flattens.
* Keep components with eigenvalues > 1 (Kaiser rule).
* Use cross-validation to see which number works best.
* Use domain knowledge to decide the minimum needed.

#**Q16)Can PCA be used for classification?**
#ANS:-
* PCA itself is not a classification method.
* It transforms data into fewer uncorrelated features (principal components).
* Then you apply a classifier (e.g., Logistic Regression, SVM, Random Forest) on these components.
* It can help improve speed, reduce overfitting, and handle multicollinearity.
* But sometimes it may reduce accuracy if important class-related variance is lost.

#**Q17)What are the limitations of PCA?**
#ANS:-
* Loss of information
* Interpretability  Principal components are combinations of original features.
* Captures only linear relationships between features.
* Features must be standardized for fair contribution.
* Outliers can distort the components.
* Ignores target labels

#**Q18)How do KNN and PCA complement each other?**
#ANS:-
* PCA reduces dimensions, making KNN faster and less affected by the curse of dimensionality.
* Removes noise & correlations, helping KNN focus on essential patterns.
* Improves distance calculations in KNN by eliminating redundant features.
* Prevents overfitting in KNN by reducing irrelevant features.
* Speeds up predictions since KNN stores fewer features per data point.

#**Q19)How does KNN handle missing values in a dataset?**
#ANS:-
* Remove rows with missing values (if few).
* Impute with mean/median/mode before applying KNN.
* KNN imputation  Replace missing values using the average of nearest neighbors for that feature.
* Use algorithms or libraries (like KNNImputer in scikit-learn) to fill in missing data first.

#**Q20)What are the key differences between PCA and Linear Discriminant Analysis (LDA)?**
#ANS:-
* Goal  PCA focuses on maximizing variance; LDA focuses on maximizing class separability.
* Supervision  PCA is unsupervised (ignores class labels); LDA is supervised (uses class labels).
* Output components  PCA can produce up to the number of original features; LDA produces at most (number of classes  1) components.
* Basis  PCA uses eigenvectors of the covariance matrix; LDA uses eigenvectors of the scatter matrices (between-class & within-class).
* Use case  PCA is for dimensionality reduction in general; LDA is mainly for classification task

```python
#Q21)Train a KNN Classifier on the Iris dataset and print model accuracy
from sklearn.datasets import load_iris
data=load_iris()
X=data.data
y=data.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
print(f"Model Accuracy:{accuracy_score(y_test,y_pred)*100:.2f}%")
```

```python
#Q23) Train a KNN Classifier using different distance metrics (Euclidean and Manhattan) and compare accuracy
n=[1,2]#1-->manhataan distance and 2--> eucliden distance
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=100,n_features=4,n_classes=2,random_state=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
for distance in n:
  clas=KNeighborsClassifier(p=distance,metric='minkowski')
  clas.fit(X_train,y_train)
  y_pred=clas.predict(X_test)
  print(f"Model Accuracy:{accuracy_score(y_test,y_pred)*100:.2f}%")
```

```python
#Q30) Train a PCA model and visualize the cumulative explained variance.

X, y = make_classification(n_samples=1000, n_features=10, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

pca = PCA()
pca.fit_transform(X_train)

cum_var = pca.explained_variance_ratio_.cumsum()

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cum_var) + 1), cum_var, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA - Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
#Q38)Train a PCA model and analyze the effect of different numbers of components on accuracy
X, y = make_classification(n_samples=500, n_features=5, n_informative=3, n_classes=3, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

components_list = range(1, X_train.shape[1] + 1)


for n in components_list:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train_pca, y_train)
    y_pred = knn.predict(X_test_pca)
    acc = accuracy_score(y_test, y_pred)
    print(f"For {n} component accuracy is: {acc*100}")
```

```python
#Q42) Train a KNN Regressor and analyze the effect of different distance metrics on prediction error
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
metrics = ['euclidean', 'manhattan', 'minkowski']
results = {}
for metric in metrics:
    knn = KNeighborsRegressor(n_neighbors=5, metric=metric)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    results[metric] = mse

for metric, mse in results.items():
    print(f"Metric: {metric}, MSE: {mse:.4f}")
```

```python
#Q46) Train a PCA model and visualize the data reconstruction error after reducing dimensions
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
n_components = 20
pca = PCA(n_components=n_components)
X_reduced = pca.fit_transform(X_scaled)
X_reconstructed = pca.inverse_transform(X_reduced)
reconstruction_error = np.mean((X_scaled - X_reconstructed) ** 2, axis=1)
plt.hist(reconstruction_error, bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Reconstruction Error (MSE)")
plt.ylabel("Frequency")
plt.title(f"Data Reconstruction Error with {n_components} Components")
plt.show()
```

```python
#Q47)Train a KNN Classifier and visualize the decision boundary
X, y = make_classification(n_samples=100, n_features=2, n_informative=2,n_redundant=0, n_classes=2, random_state=42)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_scaled, y)

x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))


Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, edgecolor='k', s=50, cmap=plt.cm.coolwarm)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("KNN Decision Boundary")
plt.show()
```
